```python
# import necessary libraries
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# in this code, we plot the ground truth data
```

```python
r_seed = 0  
i      = 8 
data   = pd.read_csv('Data/data' + str(r_seed) + '_' + str(i) + '.csv', header = None)
label  = pd.read_csv('Data/truelabel' + str(r_seed) + '_' + str(i) + '.csv', header = None)

x=data[0].to_numpy()
y=data[1].to_numpy()
color=label.to_numpy()

plt.figure(figsize=(6,6))
plt.scatter(x.reshape(-1), y.reshape(-1), s=5,c=color.reshape(-1), cmap = 'jet')
plt.show()
```
